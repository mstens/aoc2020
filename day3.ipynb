```python
with open("day1-1.txt", "r") as f:
    data = f.read()
    s = data.split("\n")
    
rows = len(s)
columns = len(s[0])

s
```

```python
r, c = 1, 3
trees_so_far = 0

while(r < rows):
    if s[r][c%columns] == "#":
        trees_so_far += 1
    elif s[r][c%columns] != ".":
        raise Exception("Ugghhh wtf marius")
    r += 1
    c += 3

print(trees_so_far)
```

```python
def get_trees(right, down, map_data=s):
    rows = len(map_data)
    columns = len(map_data[0])
    r, c = down, right
    trees_so_far = 0

    while(r < rows):
        if map_data[r][c%columns] == "#":
            trees_so_far += 1
        elif map_data[r][c%columns] != ".":
            raise Exception("Ugghhh wtf marius")
        r += down
        c += right
        
    return(trees_so_far)
```

```python
from functools import reduce

steps = [
    [1, 1],
    [3, 1],
    [5, 1],
    [7, 1],
    [1, 2]
]

trees = [get_trees(*rd) for rd in steps]
prod = reduce(lambda x,y : x*y, trees)
prod
```
